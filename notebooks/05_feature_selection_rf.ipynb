Load Feature Importance

```python
import pandas as pd
```

```python
feat_imp = pd.read_csv("../reports/rf_feature_importance.csv")
feat_imp.head()
```

Select Top-30 Features

```python
TOP_K = 30

top_features = feat_imp.head(TOP_K)["feature"].tolist()

print("Top selected features:", len(top_features))
top_features
```

Create Reduced Dataset

```python
train_df = pd.read_csv("../data/processed/nsl_kdd_train_encoded.csv")
test_df  = pd.read_csv("../data/processed/nsl_kdd_test_encoded.csv")
```

```python
X_train_fs = train_df[top_features]
y_train    = train_df["label"]

X_test_fs  = test_df[top_features]
y_test     = test_df["label"]

print(X_train_fs.shape, X_test_fs.shape)
```

Save Feature-Selected Dataset

```python
import os
os.makedirs("../data/feature_selected", exist_ok=True)

X_train_fs.to_csv("../data/feature_selected/X_train_top30.csv", index=False)
X_test_fs.to_csv("../data/feature_selected/X_test_top30.csv", index=False)
y_train.to_csv("../data/feature_selected/y_train.csv", index=False)
y_test.to_csv("../data/feature_selected/y_test.csv", index=False)

print("Feature-selected datasets saved")
```
